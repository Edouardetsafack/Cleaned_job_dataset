```python
import pandas as pd
import numpy as np
import re
```

```python
df=pd.read_csv('Uncleaned_DS_jobs.csv')
df.head()
```

```python
df=df.drop(columns='index')
```

```python
df.dtypes
```

```python
df['Salary Estimate']=df['Salary Estimate'].str.replace('(Glassdoor est.)','')
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

```python
#function which extract numbers and hyphen in Salary Estimate column
def extract_numbers(text):
    return re.sub(r'[^\d\-]',"", str(text))
df['Salary Estimate'] = df['Salary Estimate'].apply(extract_numbers)
```

```python
#We divide salary estimate in two column min_salary and max_salary
def min_max(salary):
  match = re.search(r'(\d+)-(\d+)', salary)
  if match:
    return int(match.group(1)), int(match.group(2))
  else:
    return None

df[['min_salary', 'max_salary']] = df['Salary Estimate'].apply(lambda x: pd.Series(min_max(x)))
df.head()
```

```python
newlocation=df['Location'].str.split(',',n=1,expand=True)
newlocation.rename(columns={0:'location_job',1:'job_state'})
```

```python
df[['location_job','job_state']]=df['Location'].str.split(',',n=1,expand=True)
df
```

```python
df=df.drop(columns='location_job')
```

```python
df[['Company Name','Canel']]=df['Company Name'].str.split('[0-9]',n=1,expand=True)
df.head()
```

```python
#from datetime import datetime
#df['company_age']=pd.Timestamp('now').year-df['Founded']
#df['company_age']=pd.Timestamp('2020').year-df['Founded']
#df['company_age']
Year = 2020
def companyage(year):
    if year == -1:
        return year
    else:
        age = Year - year
        return age

df['Company_age'] = df['Founded'].apply(companyage)
df
    
```

```python
df['avg_salary']=df[['max_salary','min_salary']].mean(axis=1)
```

```python
df=df.drop(columns='Founded')
```

```python
def extract_before_newline(text):
    
    if '\n' in text:
        return text.split('\n')[0]
    else:
        return text
df["Company Name"] = df["Company Name"].apply(extract_before_newline)
df.head()
```

```python
df['job_simp']=df['Job Title'].str.replace('Sr','')
df
```

```python
#df=df.drop(columns='location_job')
```

```python
def job_extraction(job):
  if 'data scientist' in job.lower():
    return 'data scientist'
  elif 'analyst' in job.lower():
    return 'analyst'
  elif 'machine learning engineer' in job.lower():
    return 'mle'
  else:
    return 'na'

df['job_simp'] = df['Job Title'].apply(job_extraction)
df
```

```python
def nomination(job):
    if 'Sr' in job:
        return 'Senior'
    else:
        return 'na'
df['Seniority'] = df['Job Title'].apply(nomination)
df    
```
